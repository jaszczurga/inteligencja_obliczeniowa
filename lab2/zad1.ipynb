```python
import pandas as pd
from sklearn.model_selection import train_test_split

INDEX_NUMBER = 286123 

df = pd.read_csv("iris_big.csv")
(train_set, test_set) = train_test_split(df.values, train_size=0.7,
random_state=INDEX_NUMBER)

def classify_iris(sl, sw, pl, pw):
    if sl < 5.4:
        return("setosa") 
    elif pl > 4.8:
        return("virginica")
    else:
        return("versicolor") 


good_predictions = 0
len_test_set = test_set.shape[0]

for i in range(len_test_set):
    current_iris = test_set[i]
    
    sl = current_iris[0]
    sw = current_iris[1]
    pl = current_iris[2]
    pw = current_iris[3]
    
    true_variety = current_iris[4]
    
    if classify_iris(sl, sw, pl, pw) == true_variety:
        good_predictions = good_predictions + 1

print("\n----------------------------------")
print(f"Liczba dobrych odpowiedzi: {good_predictions}")

accuracy_percentage = (good_predictions / len_test_set) * 100

print(f"Procentowa dokładność klasyfikatora: {accuracy_percentage:.2f}%")
print("----------------------------------")
```
