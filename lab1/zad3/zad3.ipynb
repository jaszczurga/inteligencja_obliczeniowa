```python
#A
df = df.replace("", pd.NA)
num_null_rows = df.isnull().any(axis=1).sum()
print(f"dataset has {num_null_rows} with empty values")
```

```python
#B
cols_to_convert = df.columns[:4]
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')
for col in df.select_dtypes(include='number').columns:
    mean_val = df.loc[df[col].between(0,15),col].mean()
    df[col] = df[col].where(df[col].between(0,15),mean_val)


print(df.values[120:150,:])
print(df)
```

```python
#C
from rapidfuzz import process, fuzz
valid_variety = ["Setosa","Versicolor","Virginica"]

def correct_typo(word, choices, threshold=80):
    match, score, _ = process.extractOne(word, choices, scorer=fuzz.ratio)
    if score >= threshold:
        return match
    else:
        return None 
    
df['variety'] = df['variety'].apply(lambda x: correct_typo(x, valid_variety))

print(df.values[120:150,:])
```

```python
import pandas as pd
import numpy as np
from difflib import get_close_matches

df = pd.read_csv("iris.csv")

# 1) Braki i "puste" napisy
null_counts = df.isnull().sum()
empty_string_counts = df.applymap(lambda x: isinstance(x, str) and x.strip() == "").sum()
total_missing_cells = null_counts.sum() + empty_string_counts.sum()

print("Braki null w kolumnach:\n", null_counts.to_dict())
print("Puste napisy ('' lub same spacje) w kolumnach:\n", empty_string_counts.to_dict())
print("Łączna liczba brakujących/pustych komórek:", int(total_missing_cells))

# 2) Numeryczne kolumny — wykrycie i naprawa wartości spoza (0;15)
numeric_cols = [col for col in df.columns
                if pd.to_numeric(df[col], errors='coerce').notnull().sum() > 0
                and col.lower() not in ("species", "class", "label")]
print("\nWykryte kolumny numeryczne:", numeric_cols)

out_of_range_report = {}
for col in numeric_cols:
    series = pd.to_numeric(df[col], errors='coerce')
    valid_mask = (series > 0) & (series < 15)
    # mediana obliczona na wartościach poprawnych
    median_val = series[valid_mask].median()
    # jeśli mediana jest NaN (np. brak poprawnych wartości) -> użyj mediany z wszystkich nie-NaN
    if pd.isna(median_val):
        median_val = series.dropna().median()
    replace_mask = (~valid_mask) | series.isna()
    replaced_indices = df.index[replace_mask].tolist()
    out_of_range_report[col] = {
        "invalid_count": int(replace_mask.sum()),
        "median_used": float(median_val) if not pd.isna(median_val) else None,
        "replaced_indices_sample": replaced_indices[:10]
    }
    # zastąpienie w df
    df.loc[replace_mask, col] = median_val

print("\nRaport wartości spoza (0;15) i zastąpień (przykładowe indeksy):")
for col, info in out_of_range_report.items():
    print(f" {col}: {info['invalid_count']} -> użyto mediana={info['median_used']}, przykłady indeksów {info['replaced_indices_sample']}")

# 3) Gatunki — poprawa nazw
allowed = ["Setosa", "Versicolor", "Virginica"]
allowed_lower = [a.lower() for a in allowed]

species_raw = df.get("species") if "species" in df.columns else None
if species_raw is None:
    print("\nBrak kolumny 'species' w pliku.")
else:
    cleaned = species_raw.astype(str).str.strip()
    # traktuj 'nan' string jako brak
    cleaned = cleaned.replace({"nan": np.nan})
    corrections = {}
    unknown_indices = []
    for idx, val in cleaned.items():
        if pd.isna(val) or val == "":
            cleaned.at[idx] = np.nan
            unknown_indices.append(idx)
            continue
        val_low = val.lower()
        if val_low in allowed_lower:
            # poprawna nazwa (wyprostuj kapitalizację)
            matched = allowed[allowed_lower.index(val_low)]
            if matched != val:
                corrections[idx] = (val, matched)
            cleaned.at[idx] = matched
        else:
            # spróbuj dopasować fuzzy
            cand = get_close_matches(val_low, allowed_lower, n=1, cutoff=0.6)
            if cand:
                matched = allowed[allowed_lower.index(cand[0])]
                corrections[idx] = (val, matched)
                cleaned.at[idx] = matched
            else:
                # nie znaleziono dopasowania -> oznacz jako NaN
                unknown_indices.append(idx)
                cleaned.at[idx] = np.nan

    # Wypełnij nieznane (NaN) najczęstszą nazwą (fallback)
    mode_species = cleaned.mode(dropna=True)
    mode_val = mode_species.iloc[0] if not mode_species.empty else None
    if mode_val is not None:
        cleaned = cleaned.fillna(mode_val)

    df["species"] = cleaned

    print("\nLiczba dokonanych automatycznych korekt nazw gatunków:", len(corrections))
    if corrections:
        # pokaż kilka przykładów
        sample_items = list(corrections.items())[:10]
        for idx, (orig, new) in sample_items:
            print(f" idx {idx}: '{orig}' -> '{new}'")
    print("Liczba początkowo nieznanych/nie dopasowanych indeksów (wypełnione trybem):", len(unknown_indices))
    print("Wypełniono nieznane gatunki wartością:", mode_val)

# Zapis wyniku
df.to_csv("iris_cleaned.csv", index=False)
print("\nZapisano oczyszczoną bazę: iris_cleaned.csv")
```
