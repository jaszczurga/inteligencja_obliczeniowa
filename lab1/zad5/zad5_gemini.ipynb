```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Wczytaj zbir danych Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Series(iris.target_names[iris.target], dtype='category')

# Ustaw styl wykresw
sns.set_style("whitegrid")
plt.style.use('seaborn-v0_8-whitegrid')

# Definicje kolumn do uycia
x_col = 'sepal length (cm)'
y_col = 'sepal width (cm)'

# 1. Wykres z danymi oryginalnymi
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.scatterplot(
    data=df,
    x=x_col,
    y=y_col,
    hue='species',
    palette='deep',
    s=100
)
plt.title('Dane oryginalne')
plt.xlabel('Dugo kielicha (cm)')
plt.ylabel('Szeroko kielicha (cm)')

# 2. Wykres z danymi znormalizowanymi (Min-Max)
scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[[x_col, y_col]] = scaler_minmax.fit_transform(df_minmax[[x_col, y_col]])
plt.subplot(1, 3, 2)
sns.scatterplot(
    data=df_minmax,
    x=x_col,
    y=y_col,
    hue='species',
    palette='deep',
    s=100,
    legend=False
)
plt.title('Znormalizowane (Min-Max)')
plt.xlabel('Dugo kielicha (znormalizowana)')
plt.ylabel('Szeroko kielicha (znormalizowana)')

# 3. Wykres z danymi zeskalowanymi (Z-Score)
scaler_standard = StandardScaler()
df_zscore = df.copy()
df_zscore[[x_col, y_col]] = scaler_standard.fit_transform(df_zscore[[x_col, y_col]])
plt.subplot(1, 3, 3)
sns.scatterplot(
    data=df_zscore,
    x=x_col,
    y=y_col,
    hue='species',
    palette='deep',
    s=100,
    legend=False
)
plt.title('Zeskalowane (Z-Score)')
plt.xlabel('Dugo kielicha (Z-Score)')
plt.ylabel('Szeroko kielicha (Z-Score)')

# Zakoczenie i wywietlenie wykresw
plt.tight_layout()
plt.show()

# Wywietl statystyki
print("\nStatystyki dla danych oryginalnych:")
print(df[[x_col, y_col]].describe())


#prompt 
# napisz program w pytnoie zeby rozwiazac to zadanie Stwrz wykresy z irysami jako punktami na wykresie, dla dwch zmiennych: sepal length i sepal width. Klasy irysw

# oznaczone s w legendzie wykresu. Zrb wykres w trzech wersjach: dane oryginalne, znormalizowane min-max i

# zeskalowane z-scorem. Wynik powinien przypomina ten poniej.

# Co moesz powiedzie o min, max, mean, standard deviation dla tych danych? uzywaj matplotliba skilerana etc

```
